```python
from GoogleNews import GoogleNews
import requests
from requests import HTTPError
import pandas as pd
import jinja2
import time

pd.set_option('display.max_colwidth', None)
```

```python
medias = ['La Nación', 'Clarin.com']
```

```python
googlenews = GoogleNews(lang='es', region='AR')
```

```python
search = googlenews.search('milei')
result = googlenews.result()
```

```python
def get_real_url(url: str) -> str:
    session = requests.session()

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36",
        "Accept-Language": "es-ES,es;q=0.9,en;q=0.8",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
        "Referer": "https://news.google.com/"
    }
    
    try:
        response = session.get(
            url,
            allow_redirects=True,
            timeout=30,
            headers=headers
        )
        return response.url
    except HTTPError as e:
        raise e
```

```python
all_news = []

for noticia in result:
    if noticia['media'] in medias:
        all_news.append({
           'media': noticia['media'],
           'title': noticia['title'],
            'url': noticia['link']
        })
        
        time.sleep(5)

df_news = pd.DataFrame(all_news)
df_news = df_news.drop_duplicates().reset_index(drop=True)
```

```python
df_news['url'] = df_news.apply(lambda x: get_real_url(x['url']), axis=1)
```
